```python
import pandas as pd

def add(esti, new_data):
    
   
    new_df = pd.DataFrame(new_data)
  
    combined_df = pd.concat([esti, new_df], ignore_index=True)
    return combined_df

new_data = [{
    '접수연도': 2023, '자치구코드': 11215.0, '자치구명': '광진구', '법정동코드': 10300, 
    '법정동명': '구의동', '지번구분': 1, '지번구분명': '대지', '본번': 401, '부번': 24, 
    '건물명': '구의현대2단지아파트', '계약일': 202311021, '거래금액(만원)': 15000, 
    '건물면적(m2)': 33.27, '토지면적(m2)': 33.27, '층': 18, '권리구분': '분양권', 
    '취소일': '', '건축년도': 1982, '건물용도': '아파트', '신고구분': '중개거래', 
    '신고한 공인중개사 지역명': '서울 광진구'
}]


esti = add(esti, new_data)
```

```python
#2-1.거래된 금액 기준으로 가장 높은 지역 내림차순 정리
```

```python
import pandas as pd

def highpriceregion(data):
    
    avg = data.groupby('자치구명')['거래금액(만원)'].mean()

    
    highpriceregion = data.loc[data.groupby('자치구명')['거래금액(만원)'].idxmax(), ['자치구명', '건물명', '층', '거래금액(만원)']]

 
    highpriceregion['평균 거래금액'] = highpriceregion['자치구명'].map(avg)  # 평균 거래금액을 합치기
    sorted_data = highpriceregion.sort_values(by='평균 거래금액', ascending=False)  # 내림차순 정렬

    
    print("\n지역별 거래 정보")
    for idx, row in sorted_data.iterrows():
        print(f"- 지역: {row['자치구명']}")
        print(f"  평균 거래금액: {row['평균 거래금액']:.2f}만원")
        print(f"  최고 거래가 건물명: {row['건물명']}, 층수: {int(row['층']) if not pd.isna(row['층']) else '정보 없음'}, 거래금액: {row['거래금액(만원)']}만원\n")


highpriceregion(esti)
```

```python
#2-2.지역, 건물용도 입력받아 높은 가격부터 내림차순 정리
```

```python
def highprice(data):
    
    region_name = input("지역명을 입력하세요: ")
    
   
    building_type = input("건물용도를 입력하세요 (예: 아파트, 오피스텔, 연립다세대): ")
    
    filtered_data = data[(data['자치구명'] == region_name) & (data['건물용도'] == building_type)]
    
   
    if filtered_data.empty:
        print(f"'{region_name}'에 '{building_type}'에 해당하는 데이터가 없습니다.")
        return
    
    
    sorted_data = filtered_data.sort_values(by='거래금액(만원)', ascending=False)
    
    print(f"\n'{region_name}'의 '{building_type}' 중 높은 가격의 건물명 (내림차순):")
    for idx, row in sorted_data.iterrows():
        print(f"- 건물명: {row['건물명']}, 거래금액: {row['거래금액(만원)']}만원")


highprice(esti)
```

```python
#2-3.지역별 거래량
```

```python

import pandas as pd
import matplotlib.pyplot as plt


region_counts = esti['자치구명'].value_counts()


plt.figure(figsize=(12, 6))
region_counts.sort_values(ascending=False).plot(kind='bar', color='skyblue', edgecolor='black')


plt.title('지역별 거래량', fontsize=16)
plt.xlabel('자치구명', fontsize=12)
plt.ylabel('거래 건수', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.tight_layout()  
plt.show()
```
